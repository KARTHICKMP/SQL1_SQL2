```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", names=["age","workclass","fnlwgt","education","education-num","marital-status","occupation","relationship","race","sex","capital-gain","capital-loss","hours-per-week","native-country","weighted_tallies"], header=None)
```

###### 1. Create an sqlalchemy engine using a sample from the data set

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)
df.to_sql('adult', con=engine)
engine.execute("SELECT * FROM adult limit 5").fetchall()
```

###### 2. Write two basic update queries

```python
engine.execute("UPDATE adult SET workclass = 'State-government' WHERE education = ' Bachelors'")
engine.execute("SELECT * FROM adult limit 5").fetchall()
```

```python
engine.execute("UPDATE adult SET workclass = 'State-govern' WHERE education = ' Bachelors'")
engine.execute("SELECT * FROM adult limit 5").fetchall()
```

###### 3. Write two delete queries

```python
engine.execute("DELETE FROM adult WHERE education = ' Bachelors'")
engine.execute("SELECT * FROM adult limit 5").fetchall()
```

```python
engine.execute("DELETE FROM adult WHERE workclass = ' Private'")
engine.execute("SELECT * FROM adult limit 5").fetchall()
```

###### 4. Write two filter queries

```python
engine.execute("SELECT * FROM adult WHERE age LIKE '4%'").fetchall()
```

```python
engine.execute("SELECT * FROM adult WHERE 'native-country' LIKE '%-%'").fetchall()
```

###### 5. Write two function queries

```python
res = engine.execute("SELECT count(*) FROM adult")
for i in res:
    print(i)
```

```python
res = engine.execute("SELECT max(age) FROM adult")
for i in res:
    print(i)
```
